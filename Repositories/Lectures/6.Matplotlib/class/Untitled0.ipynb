```python
figure(figsize(9.5, 9))
scatter(x, y)
plot(X, Y, linestyle="--", linewidth=3.5, c='g')
plot(2*X, Y, linestyle="-.", linewidth=2.5, c='r')
xlim([20, 60])
ylim([0, 1000])
tick_params(axis="both", labelsize=20)
xlabel("$X\ data$", fontsize=35)
ylabel("$\mathrm{Y^2 = x}$", fontsize=35)
title("$Mi\ primer\ plot$", fontsize=30)
```
