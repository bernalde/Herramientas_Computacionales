```python
%pylab inline

f = open("data1.dat", "w")

x = linspace(0, 10, 100)
y = 10*e**(x-3)
dy = randn(y.size)*1000
y1 = y+dy
scatter(x, y1)

for i in range(len(x)):
    f.write(str(x[i]) + " " + str(y1[i])+ "\n")
f.close()	



f = open("data2.dat", "w")

x = linspace(0, 10, 100)
y = x**2 + 7
dy = randn(y.size)*10
y1 = y+dy
scatter(x, y1)

for i in range(len(x)):
    f.write(str(x[i]) + " " + str(y1[i])+ "\n")
f.close()

f = open("data3.dat", "w")

x = linspace(0, 10, 100)
y = -2*x - 7
dy = randn(y.size)*5
y1 = y+dy
scatter(x, y1)

for i in range(len(x)):
    f.write(str(x[i]) + " " + str(y1[i])+ "\n")
f.close()


f = open("data4.dat", "w")

x = linspace(0, 3*pi, 100)
y = sin(2*x)**2
dy = randn(y.size)*0.1
y1 = y + dy
scatter(x, y1)

for i in range(len(x)):
    f.write(str(x[i]) + " " + str(y1[i])+ "\n")
f.close()
```
