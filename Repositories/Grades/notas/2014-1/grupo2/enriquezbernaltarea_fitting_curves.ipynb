```python
data_3=loadtxt('data3.dat')
```

```python
x=data_3[:,0]
```

```python
y=data_3[:,1]
```

```python
scatter(x,y)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
```

```python
def coeficiente_de_correlacion (xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
from scipy.optimize import curve_fit
```

```python
def linefunc(x,m,b):
    return m*x + b
```

```python
fitpars, covmat = curve_fit(linefunc,x,y)
yteo  = x*fitpars[0] + fitpars[1]

R2=coeficiente_de_correlacion(y, yteo)

plot(x,linefunc(x,fitpars[0], fitpars[1]), 'r-', linewidth='2.0', label="Fit")
scatter(x,y)
legend(loc='best')
text(2,1,"R2="+str(R2),fontsize=10)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
```
