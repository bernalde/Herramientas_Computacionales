```python
def monedas(N_trials, M):
    j = 0
    resultados = []
    while j < M:
        j +=1
        i = 0
        caras = 0
        sellos = 0
        while i < N_trials:
            i +=1
            resultado = np.random.randint(0, 2)       
            
            if resultado == 0:
                caras +=1
            else:
                sellos +=1
                
        
        fcaras = caras/N_trials
       
        resultados.append(fcaras)
       
            
    return resultados
                
            


        
        
        
```

```python
resultados = monedas(1000.0,10000.0)
print resultados
```

```python
hist(resultados, bins=20)
xlabel("Fraccion de Caras")
```

```python
print "El intervalo de confianza de 0.1 está en",0.1/np.std(resultados), "sigmas"
print "El intervalo de confianza de 0.35 está en",0.35/np.std(resultados), "sigmas"
print "El intervalo de confianza de 0.4 está en",0.4/np.std(resultados), "sigmas"
print "El intervalo de confianza de 1.7 está en",1.7/np.std(resultados), "sigmas"
```
