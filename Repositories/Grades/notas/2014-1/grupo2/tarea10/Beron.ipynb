```python
i=15
j=1.1

x = np.random.rand(100000)*(2*i) - i
y = np.random.rand(100000)*j
scatter(x,y)
```

```python
index = np.where(y<=exp(-x**2))
```

```python
xc=x[index]
yc=y[index]
```

```python
scatter(x,y, c='g')
scatter(xc,yc, c='r')
```

```python
A = len(xc)  # Numero de puntos bajo de la funcion.
B = len(x)   # Numero de puntos totales. 
```

```python
result = j*(2*i)*float(A)/float(B)
print "Integral= ", result
```
