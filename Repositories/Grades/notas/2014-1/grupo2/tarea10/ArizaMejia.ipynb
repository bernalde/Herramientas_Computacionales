```python
import numpy as np
```

```python
x = 6*(np.random.rand(100000)- 0.5)
y = np.random.rand(100000)*1
```

```python
x
```

```python
y
```

```python
index = np.where(y<np.exp(-(x**2)))
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(9,9))
scatter(x,y, c='g')
scatter(xc,yc, c='r')
```

```python
X = len(xc)
Y = len(x)
area = (max(x)-min(x))*(max(y)-min(y))
sol = area * (float(X)/float(Y))
```
