```python
x = np.random.rand(100000)*5.2 -2.5
y = np.random.rand(100000)*(3/2)
```

```python


figure(figsize=(9, 9))
scatter(x, y, alpha=0.5, s=10)

```

```python


index = np.where(y<e**(-x**2))
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(12,9))
scatter(x,y, c='g')
scatter(xc,yc, c='r')
```

```python
X =len (xc)
Y= len (x)
```

```python
h = float(X)/float(Y) 
punto = h*5.2*(3/2)
```
