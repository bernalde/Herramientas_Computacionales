```python
a=10
b=2

x = np.random.rand(100000)*2*a - a
y = np.random.rand(100000)*b
```

```python
figure(figsize=(9, 9))
scatter(x, y, alpha=0.2, s=10)
```

```python
index = np.where(y<=exp(-(x**2)))
```

```python
xe=x[index]
ye=y[index]
figure(figsize=(9,9))
scatter(x,y, c='g')
scatter(xe,ye, c='r')
```

```python
X = len(xe)  # Numero de puntos dentro del area
Y = len(x)   # Numero de puntos totales. 
areac=2*a*b
integral = areac* float(X)/float(Y)
print "Integral = ", integral
```
