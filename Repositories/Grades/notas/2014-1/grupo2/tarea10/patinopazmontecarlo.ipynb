```python
x = np.random.rand(1000000)*5 - 2.5
y = np.random.rand(1000000)
```

```python
scatter(x,y)
```

```python
index = np.where(y<e**(-x**2))
xc = x[index]
yc = y[index]
```

```python
scatter(x,y, c='g')
scatter(xc,yc, c='r')
```

```python
X = len(xc)
Y = len(x)
area = ((max(x)-min(x))*(max(y)-min(y)))
integral = area*float(X)/float(Y)
print integral
```
