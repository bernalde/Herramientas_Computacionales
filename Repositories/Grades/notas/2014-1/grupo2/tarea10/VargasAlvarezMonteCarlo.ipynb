```python
b = 5
h = 1
```

```python
x = random.rand(100000)*b - b/2.0
y = random.rand(100000)*h
```

```python
i = where(y < exp(-x**2))
```

```python
xf = x[i]
yf = y[i]
```

```python
figure(figsize=(13,13))
scatter(x, y, c="g")
scatter(xf, yf, c="y")
```

```python
c = len(x)
m = len(xf)
Ac = b*h
Am = (Ac/float(c))*float(m)
error = Am - sqrt(pi)
```
