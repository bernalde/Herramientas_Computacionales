```python
x = np.random.rand(100000)*8 - 4
y = np.random.rand(100000)*1
```

```python
figure(figsize=(8, 8))
scatter(x, y, alpha=0.2, s=10, c= 'grey')
```

```python
index = np.where(y<e**(-x**2))
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(8, 8))
scatter(x,y, c='cyan')
scatter(xc,yc, c='purple')
```

```python
X = len(xc)
Y = len(x)
area = (max(x)-min(x))*(max(y)-min(y))
```

```python
areafunc = area* float(X)/float(Y)
print "integral = ", areafunc
```
