```python
x = np.random.rand(10000)*6- 3
y = np.random.rand(10000)*1 - 0

```

```python
figure(figsize=(9, 9))
scatter(x, y, alpha=0.2, s=10)

```

```python
index = np.where(y<(e**(-x**2)))
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(9,9))
scatter(x,y, c='cyan')
scatter(xc,yc, c='black')
```

```python
X = len(xc)
Y = len(x) 
area=(max(x)-min(x))*(max(y)-min(y))
```

```python
i = area*(float(X)/float(Y))
print "integral = ", i
```
