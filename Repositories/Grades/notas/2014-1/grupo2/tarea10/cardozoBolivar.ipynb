```python
x = np.random.rand(100000)*10 - 5
y = np.random.rand(100000)*1
```

```python
figure(figsize=(9,9))
scatter(x,y, alpha =0.2, s=10)
```

```python
index = np.where(y<e**-x**2)
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize(9,9))
scatter(x,y, c ='g')
scatter(xc,yc, c='r')
```

```python
X = len(xc)
Y = len(x)
areat = (max(x)-min(x))*(max(y)-min(y))
```

```python
area = areat*float(X)/float(Y)
print 'integral = ', area
```
