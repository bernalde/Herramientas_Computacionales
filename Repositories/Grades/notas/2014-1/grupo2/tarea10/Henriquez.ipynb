```python
x=np.random.rand(100000)*6-3
y=np.random.rand(100000)
```

```python
figure(figsize=(9, 9))
scatter(x, y, alpha=0.2, s=10)
```

```python
index=np.where(exp(-x**2)<y)
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(9,9))
scatter(x,y, c='b')
scatter(xc,yc, c='r')
```

```python
X = len(xc)  
Y = len(x)   
```

```python
integral =  2*(log(2))**(0.5)*(float(X)/float(Y))
print "integral = ", integral
```
