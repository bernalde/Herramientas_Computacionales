```python
def funcion(x):
    y= (exp(-(x**2)))/(((x-3)**2)+(0.01**2))
    return y
```

```python
X = linspace(-2, 4, 100)
Y = funcion(X)

plot(X,Y)
```

```python
walk = []
x = random.random()*4
walk.append(x)

```

```python
for i in range(200000):
    x = random.random()*2-1 
    alpha = funcion(x + walk[-1])/funcion(walk[-1])

    if alpha>=1.0:
        walk.append(walk[-1]+x)
    else:
        beta = random.random()
        if(beta<=alpha):
            walk.append(walk[-1]+x)
        else:
            walk.append(walk[-1])
    
```

```python

histo = hist(walk, bins=40, color = 'green', normed=True)
f = funcion(X)
norma = sum(f*(X[1]-X[0]))
ylim(0,0.5)
plot(X,f/norma, linewidth=1, color='r')

```
