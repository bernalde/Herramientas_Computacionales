```python
import matplotlib.pyplot as pl
%pylab inline
```

```python
data=genfromtxt('data.txt')
```

```python
t=data[:,2]
v=data[:,3]
w=data[:,4]
x=data[:,5]
y=data[:,6]
z=data[:,7]


print x
```

```python
figure(figsize=(20, 16))
subplots_adjust(hspace=0.00, wspace=0.00)

subplot(321)
scatter(t, v, c='y')
xlabel("$t$", fontsize=30)
ylabel("$v$", fontsize=30)

subplot(322)
scatter(v, w, c='y')
xlabel("$v$", fontsize=30)
ylabel("$w$", fontsize=30)

subplot(323)
scatter(w, x, c='b')
xlabel("$w$", fontsize=30)
ylabel("$x$", fontsize=30)

subplot(324)
scatter(t, z, c='b')
xlabel("$t$", fontsize=30)
ylabel("$z$", fontsize=30)

subplot(325)
plot(x, y, c='r')
xlabel("$x$", fontsize=30)
ylabel("$y$", fontsize=30)

subplot(326)
plot(y, z, c='r')
xlabel("$y$", fontsize=30)
ylabel("$z$", fontsize=30)
```

```python
figure(figsize=(25,12))
subplots_adjust(hspace=0.00, wspace=0.00)

subplot (131)
tick_params(axis="both", labelsize=20)
hist2d(x, y, bins=(50, 50))
colorbar()
xlabel("$x$", fontsize=30)
ylabel("$y$", fontsize=30)
subplot (132)
tick_params(axis="both", labelsize=20)
hist2d(v, w, bins=(50, 50))
colorbar()
xlabel("$v$", fontsize=30)
ylabel("$w$", fontsize=30)
subplot (133)
tick_params(axis="both", labelsize=20)
hist2d(t, z, bins=(50, 50))
colorbar()
xlabel("$t$", fontsize=30)
ylabel("$z$", fontsize=30)
```

```python


figure(figsize=(20, 16))
subplots_adjust(hspace=0.00, wspace=0.00)

subplot(321)
tick_params(axis="both", labelsize=20)
hist(t, bins=30)
xlabel("$t$", fontsize=30)
ylim([0, 3*max(t)])

subplot(322)
tick_params(axis="both", labelsize=20)
hist(v, bins=30)
xlabel("$v$", fontsize=30)
ylim([0, 3*max(v)])

subplot(323)
tick_params(axis="both", labelsize=20)
hist(w, bins=30)
xlabel("$w$", fontsize=30)
ylim([0, 3*max(w)])

subplot(324)
tick_params(axis="both", labelsize=20)
hist(x, bins=30)
xlabel("$x$", fontsize=30)
ylim([0, 3*max(x)])

subplot(325)
tick_params(axis="both", labelsize=20)
hist(y, bins=30)
xlabel("$y$", fontsize=30)
ylim([0, 3*max(y)])

subplot(326)
tick_params(axis="both", labelsize=20)
hist(z, bins=30)
xlabel("$z$", fontsize=30)
ylim([0, 3*max(z)])
```
