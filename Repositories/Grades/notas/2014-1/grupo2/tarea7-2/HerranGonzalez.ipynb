```python
def carasello(intentos,mediciones):
    i=0
    j=0
    Caras = 0
    Sellos = 0
    
    while i < mediciones:
                    i+=1
                    while j < intentos:
                        j+=1
                        x = random.randint(1,3)
                        if x==1:
                            Caras+=1
                            
                        elif x==2:
                            Sellos+=1
                                     
    return Caras, Sellos


#Funcion 'carasello()' lanzala moneda(intentos) y eso lo repite ciertas veces (mediciones).....- El metodo retorna el numero de caras y sellos que salieron
carasello(1000.0,20.0)
```

```python
def recorridoResta():
    i=0
    resta = []
    lanzamientos = []
    while i < 20:
         i+=1   
         x, y = carasello(1000.0,20.0)
         resta.append(abs(mean(x)-mean(y)))  
         lanzamientos.append(i*1000.0)
    return lanzamientos, resta

            
#De acuerdo al punto 1 de la tarea.... este metodo repite 20 veces el metodo anterior y con los datos que retorna saca un promedio de caras y un promedio de sellos para luego restarlos
#Retorna en numero de lanzamientos y la resta de cada lanzamiento
```

```python
def recorrido1():
    i=0
    resta = []
    lanzamientos = []
    while i < 40:
         i+=1   
         x, y = carasello(1000.0,40.0)
         resta.append(abs(mean(x)-mean(y)))  
         lanzamientos.append(i*1000.0)
    return lanzamientos, resta
            
#De acuerdo al punto 1 de la tarea.... este metodo repite 40 veces el metodo anterior y con los datos que retorna saca un promedio de caras y un promedio de sellos para luego restarlos
#Retorna en numero de lanzamientos y la resta de cada lanzamiento
```

```python
figure(figsize=(7,7))

x,y= recorridoResta()
scatter(x,y, c='g', s = 65)

m,n = recorrido1()
scatter(m,n, c= 'r',marker = '+',  s =80)

xlabel("Numero de Lanzamientos", fontsize =15, fontweight= 'bold')
ylabel("Resta(caras-sellos)", fontsize= 15, fontweight= 'bold')
title("Diferencia para 20 y 40 lanzamientos", fontsize=25, fontstyle= 'oblique')

#Graficar los dos metodos anteriores en un mismo scatter para comparar los resultados 
#Podemos ver que los resultados de los 40 lanzamientos se encuentran un poco mas dispersos que los de 20
```
