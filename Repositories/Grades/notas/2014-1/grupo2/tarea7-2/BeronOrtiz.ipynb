```python
def carasellaso (N, M):
    i=0
    caras = []
    sellos = []
    while i < M:
        c = 0
        s = 0
        j = 0
        while j < N:
            x = np.random.randint(0,2)
            if (x == 0):
                c +=1
            elif x==1:
                s +=1
            j+=1
        caras.append(c)
        sellos.append(s)
        i+=1
    return caras, sellos
```

```python
def coinflip (M):    
    caras = 0
    sellos = 0
    for i in range (M):
        x = np.random.randint(0,2)
        if x == 0:
            caras +=1
        elif x ==1 :
            sellos +=1
    return caras, sellos
```

```python
lanzamientos = []
diferencia = []
for i in range (20):
    x, y = carasellaso(i+1, 20)
    a= x[i]
    b= y[i]
    diferencia.append(abs(a-b))
    lanzamientos.append(i+1)
scatter(lanzamientos, diferencia)
xlabel("Lanzamientos")
ylabel("Caras - Sellos")
title("20 Lanzamientos", fontsize=20)
```
