```python
def fracciones(N, M):
    i=0
    fraccion=[]
    while i<M:
        j=0
        i+=1
        cara=0
        while j < N:
            j+=1 
            x= np.random.randint(1,3)
            if x==1:
                cara+=1
        f=cara/N
        fraccion.append(f)
    return fraccion
```

```python
fraccion= fracciones(100.0, 100.0)
```

```python
def diferencia(N,M):
    i=0
    lstcaras=[]
    lstsellos=[]
    while i<M:
        j=0
        i+=1
        caras=0
        sellos=0
        while j<N:
            j+=1
            x=np.random.randint(1,3)
            if x==1:
                caras+=1
            if x==2:
                sellos+=1
        lstcaras.append(caras)
        promcaras=np.mean(lstcaras)
        lstsellos.append(sellos)
        promsellos=np.mean(lstsellos)
        diferencias= abs(promcaras-promsellos)
    return diferencias
```

```python
def puntos(M):
    i=1000
    puntos=[]
    lanzamientos=[]
    while i<20001:
        i+=1000
        procedimiento= diferencia(i,M)
        puntos.append(np.mean(procedimiento))
        lanzamientos.append(i)
    return lanzamientos, puntos
```

```python
x, y= puntos(20)
a, b = puntos(40)
plt.scatter(x,y)
plt.scatter(a,b, c='red')
plt.title("Diferencia")
plt.xlabel("Cantidad de lanzamientos")
plt.ylabel("Diferencias")
```
