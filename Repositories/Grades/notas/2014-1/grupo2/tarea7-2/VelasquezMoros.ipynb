```python
def fracciones(N, M):
    i=0
    fracciones = []
    while i < M:
        j = 0
        i+=1
        caras=0 
        while j < N:
            j +=1
            x = np.random.randint(1, 3)
            if x ==1:
                caras+=1
        f = caras/N
        fracciones.append(f)
        
    return fracciones
```

```python
fraccion = fracciones(100.0, 100.0)
```

```python
def diferencia(N, M):
    i=0
    lstcaras = []
    lstsellos = []
    while i < M:
        j = 0
        i+=1
        caras=0 
        sellos=0
        while j < N:
            j +=1
            x = np.random.randint(1, 3)
            if x ==1:
                caras+=1
            if x ==2:
                sellos+=1
        lstcaras.append(caras)
        promcaras = np.mean(lstcaras)
        lstsellos.append(sellos)
        promsellos = np.mean(lstsellos)
        
        diferencias = abs(promcaras-promsellos)
        
    return diferencias  
```

```python
def puntosdif(M):
    i=1000
    puntos = []
    lanzamientos = []
    while i < 20001:
        i+=1000
        diferencias = diferencia(i, M)
        puntos.append(np.mean(diferencias))
        lanzamientos.append(i)
    return lanzamientos, puntos
```

```python
x, y = puntosdif(20)
a, b = puntosdif(40)
plt.scatter(x, y, c= 'orange')
plt.scatter(a, b, c='yellow')
plt.title("Diferencias", size = 17)
plt.xlabel("$Numero$ $de$ $lanzamientos$")
plt.ylabel("$Diferencias$ $de$ $caras-sellos$")
```
