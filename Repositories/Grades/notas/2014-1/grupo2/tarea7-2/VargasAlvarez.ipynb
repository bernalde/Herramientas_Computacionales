```python
def moneda(N):
    n = 0
    caras = 0.0
    while(n < N):
        n = n + 1
        x = random.randint(0, 2)
        if(x == 0):
            caras = caras + 1
    return caras/N
```

```python
def moneda2(N, M):
    n = 0
    caras = []
    sellos = []
    while(n < M):
        n = n + 1
        m = moneda(N)
        caras.append(m*N)
        sellos.append((1 - m)*N)
    y = abs(mean(caras) - mean(sellos))
    return y
```

```python
def ejex(P):
    n = 0
    x = []
    while(n < P):
        x.append((n+1)*1000)
        n = n + 1
    return x
```

```python
def ejey(M, P):
    n = 0
    y = []
    while(n < P):
        y.append(moneda2((n+1)*1000, M))
        n = n + 1
    return y
```

```python
n = 10
m1 = 20
m2 = 40
figure(figsize = (10, 10))
v = ejex(n)
w = ejey(m1, n)
x = ejex(n)
y = ejey(m2, n)
scatter(v, w, c="b")
scatter(x, y, c="r")
title("punto 1A para 20(azul) y 40(rojo) mediciones", fontsize = 30)
xlabel("lanzamientos", fontsize = 20)
ylabel("abs(caras-sellos)", fontsize = 20)
```

```python
def moneda3(N, M):
    n = 0
    caras = []
    sellos = []
    while(n < M):
        n = n + 1
        m = moneda(N)
        caras.append(m*N)
        sellos.append((1 - m)*N)
    y = mean(caras)/mean(sellos)
    return y
```

```python
def ejeyB(M, P):
    n = 0
    y = []
    while(n < P):
        y.append(moneda3((n+1)*1000, M))
        n = n + 1
    return y
```

```python
n = 50
m = 40
figure(figsize = (10, 10))
x = ejex(n)
y = ejeyB(m, n)
scatter(x, y, c="g")
title("punto 1B", fontsize = 30)
xlabel("lanzamientos", fontsize = 20)
ylabel("caras/sellos", fontsize = 20)
```
