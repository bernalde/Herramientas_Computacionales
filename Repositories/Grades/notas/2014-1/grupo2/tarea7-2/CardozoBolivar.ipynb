```python
def coinflip(N):  #Esta funcion me produce la cantidad de cara y sello de N lanzamientos
    cara = 0
    sello = 0
    i=0
    while i < N:
        x = np.random.randint(0, 10)/5.0
        if x >= 1.0:
            cara+=1
        elif x<1.0:
            sello+=1
        i+=1
    return cara, sello
```

```python
def Medida(n,m):  #Esta función de Medida, realizara m medidas de n lanzamientos de moneda, observar que devolvera la diferencia cara sello 
    caras = []
    sellos = []
    i = 0 
    while i < m:
        x,y = coinflip(n)
        caras.append(x)
        sellos.append(y)
        i = i + 1
    return np.abs( mean(caras) - mean(sellos))
        
```

```python
def data(m):   #Esta función, sera mi recoleción de las medidas en vectores, cuya longitud seran de m elementos
    x = []
    y = []
    i = 0
    while i < m:
        y.append(Medida(1000 + 500*i, m))
        x.append(1000 + 500*i)
        i = i + 1
    return x,y
        
        
```

```python
import matplotlib.pyplot as plt
```

```python
a,b = data(20);
```

```python
figure(figsize=(7, 7))
plt.scatter(a,b)
xlabel("Numero de lanzamientos", fontsize=15)
ylabel("$\mid {Cara} - {Sello} \mid $", fontsize=25)
title("$\mathrm{Figura\ con\ 20\ medidas}$", fontsize=35)
```

## Promedio de abs(cara-sello) en funcion del numero de lanzamiento en escala logaritmica.

```python
def data(m):   #Modificacion de data para promedio. Esta función, sera mi recoleción de las medidas en vectores, cuya longitud seran de m elementos
    x = []
    y = []
    i = 0
    while i < m:
        z = (Medida(1000 + 500*i, m) + sum(y) ) / (i+1)  #def de promedio
        y.append(z) #Para producir mi promedio
        x.append(1000 + 500*i)
        i = i + 1
    return x,y
        
```

## Repita el punto anterior pero esta vez con la desviación estandar.

```python
def Medida(n,m):  #Por cuarta vez, redefinamos Medida
    caras = []
    sellos = []
    i = 0 
    while i < m:
        x,y = coinflip(n)
        caras.append(x)
        sellos.append(y)
        i = i + 1
    return np.abs( mean(caras) - mean(sellos)) #La modificacion aca es para que me devuelva la diferencia
```

```python
def data(m):   #Modificacion de data para desviacion estandar. Esta función, sera mi recoleción de las medidas en vectores, cuya longitud seran de m elementos
    x = []
    y = []
    i = 0
    while i < m:
        if i != 0:
            y.append(Medida(1000 + 500*i, m))
            z = np.std(y) #saco std de la lista que llevo.
            y[i] = z #remplazo el dato del append de arriba por el std que saqué.
        if i == 0:
            y.append(Medida(1000 + 500*i, m)) #para por lo menos tener 2 datos antes de sacar el std 
        x.append(1000 + 500*i)
        i = i + 1
    return x,y
```

```python
stdx, stdy = data(60)
```

```python
figure(figsize=(7, 7))
plt.scatter(stdx,stdy)
xlabel("Numero de lanzamientos", fontsize=15)
ylabel("$ Desviacion \mid {Cara} - {Sello} \mid $", fontsize=25)
title("$\mathrm{Figura\ con\ 60\ medidas}$", fontsize=20)
plt.xscale('log')
plt.yscale('log')
```
