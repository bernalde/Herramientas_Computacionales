```python
def monedas(N_trials, M):
    j = 0
    caras = []
    sellos = []
    while j < M:
        j +=1
        i = 0
        cara = 0
        sello = 0
        while i < N_trials:
            i +=1
            resultado = np.random.randint(0, 2)       
            
            if resultado == 0:
                cara +=1
            else:
                sello +=1
                
        
        fcaras = cara/N_trials
        fsellos = sello/N_trials
       
        caras.append(cara)
        sellos.append(sello)
        
       
            
    return caras, sellos
```

```python
datos1 = []
intentos1 = []
i = 0
while i < 20:
    i+=1
    x, y = monedas(10, 100*i)
    resta = abs(mean(x) - mean(y))
    datos1.append(resta)
    intentos1.append(10*(100*i))


```
