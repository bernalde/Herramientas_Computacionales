```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def coinflip(N):
    cara = 0
    sello = 0
    i=0
    while i < N:
        x = np.random.randint(0, 10)/5.0
        if x >= 1.0:
            cara+=1
        elif x<1.0:
            sello+=1
        i+=1
    return cara, sello
```

```python
def diferenciasYRazones(medidas):
    numLanzamientos = []
    diferencias = []
    razones = []
    for i in range(20):
        lanzamientos = 1000 + (1000 * i)
        caras = []
        sellos =[]
        intentos = []
        for j in range(medidas):
            x,y = coinflip(lanzamientos)
            caras.append(x)
            sellos.append(y)
            intentos.append(j)
        promCaras = mean(caras)
        promSellos = mean(sellos)
        diferenciaAbs = abs(promCaras - promSellos)
        razon = promCaras/promSellos
        diferencias.append(diferenciaAbs)
        razones.append(razon)
        numLanzamientos.append(lanzamientos)
    return numLanzamientos,diferencias,razones
```

```python
def diferenciasYRazonesPromedio(medidas):
    numLanzamientos = []
    promediosDiferencias = []
    promediosRazones = []
    for i in range(20):
        diferenciasAbsolutas = []
        razones = []
        lanzamientos = 1000 + (1000 * i)
        for j in range(medidas):
            x, y = coinflip(lanzamientos)
            diferenciasAbsolutas.append(abs(x - y))
            razones.append(x/y)
        promediosDiferencias.append(mean(diferenciasAbsolutas))
        promediosRazones.append(mean(razones))
        numLanzamientos.append(lanzamientos)
    return numLanzamientos, promediosDiferencias, promediosRazones
```

```python
def diferenciasYRazonesDesviacion(medidas):
    numLanzamientos = []
    desviacionesDiferencias = []
    desviacionesRazones = []
    for i in range(20):
        diferenciasAbsolutas = []
        razones = []
        lanzamientos = 1000 + (1000 * i)
        for j in range(medidas):
            x, y = coinflip(lanzamientos)
            diferenciasAbsolutas.append(abs(x - y))
            razones.append(x/y)
        desviacionesDiferencias.append(std(diferenciasAbsolutas))
        desviacionesRazones.append(std(razones))
        numLanzamientos.append(lanzamientos)
    return numLanzamientos, desviacionesDiferencias, desviacionesRazones
```
