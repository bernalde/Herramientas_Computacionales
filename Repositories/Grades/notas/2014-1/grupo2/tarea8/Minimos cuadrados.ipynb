```python
data4 = np.loadtxt("data4.dat", delimiter = ' ')
```

```python
g = data4[0:, 0]
h = data4[0:, 1]
plt.scatter(g, h)
```

```python
def sinfunc(x,a,b):
    return a*np.sin(4*x-b)+0.5
fitpars, covmat = curve_fit(sinfunc,g,h)

plot(g, sinfunc(g, fitpars[0], fitpars[1]), 'r-')
scatter(g, h)
xlabel("x")
ylabel("y")
```

```python
yteo4 = fitpars[0]*(np.sin(4*g-fitpars[1]))+0.5
print "La ecuacion es " + str(fitpars[0]) + "*sin(4x - " + str(fitpars[1]) + ") + 0.5"
print "El coeficiente de correlacion es " + str(R2(h, yteo4))
```
