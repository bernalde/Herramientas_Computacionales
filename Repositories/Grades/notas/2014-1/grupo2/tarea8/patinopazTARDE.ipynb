```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
data1 = loadtxt("data1.dat")
data2 = loadtxt("data2.dat")
data3 = loadtxt("data3.dat")
data4 = loadtxt("data4.dat")
```

```python
def func(x,a,b):
    return a*e**(x*b)
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2

x = data1[:,0]
y = data1[:,1]
fitpars, covmat = curve_fit(func,x,y)
yteo  = fitpars[0]*e**(x*fitpars[1])

plot(x, func(x, fitpars[0], fitpars[1]), 'r-', label=R2(y, yteo) )
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("Data 1", fontsize=25)
legend(loc="best")
scatter(data1[:,0],data1[:,1])


print "El coeficiente de determinacion es"
print "Se uso la ecuacion y =",fitpars[0],"exp(",fitpars[1],"x)"
```

```python
def func(x,a,b,c):
    return a*x**2+b*x+c
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2

x2 = data2[:,0]
y2 = data2[:,1]
fit =  polyfit(x2, y2, 2)
yteo  = fit[0]*x**2 + fit[1]*x + fit[2]


plot(x, func(x, fit[0], fit[1],fit[2]), 'r-', label=R2(y2, yteo))
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("Data 2", fontsize=25)
legend(loc="best")
scatter(x2,y2)


print "El coeficiente de determinacion es",R2(y2, yteo) 
print "Se uso la ecuacion y =", fit[0],"*x²", fit[1],"*x+", fit[2]
```
