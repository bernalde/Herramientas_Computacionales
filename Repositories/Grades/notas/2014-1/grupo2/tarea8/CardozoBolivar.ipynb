```python
from scipy.optimize import curve_fit
```

```python
x = data[:,0]
y = data[:,1]
def expfunc(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(expfunc,x,y)
```

```python
plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(x, y)
```

#### Para esta aproximacion de curva, la ecuacion de la funcion es: 1.68251679*e**(x-1.19179788) con un R2 de 0.8486151928093747

```python
fitpars
```

```python
def xteoricos(x):
    y = fitpars[0]*e**(x-fitpars[1])
    return y
```

```python
yteoricos = xteoricos(data[:,0])
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
R2(data[:,1],yteoricos)
```
