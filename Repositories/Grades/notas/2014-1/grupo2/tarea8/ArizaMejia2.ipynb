```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
datos1 = np.loadtxt("data1.dat", delimiter=' ')
```

```python
x, y = datos1[0:,0], datos1[0:,1]
plt.scatter(x,y)
```

```python
from scipy.optimize import curve_fit
```

```python
def expfunc(x,a,b):
    return a*e**(x-b)
```

```python
fitpars, covmat = curve_fit(expfunc,x,y)
```

```python
def linfunc(x,a,b):
    return a*x+b
```

```python
x, y = datos1[0:,0], datos1[0:,1]
plt.scatter(x,y)
fitpars, covmat = curve_fit(expfunc,x,y)
plt.plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-')
```

```python
datos2 = np.loadtxt("data2.dat", delimiter=' ')
```

```python
x, y = datos2[0:,0], datos2[0:,1]
plt.scatter(x,y)
```

```python
x, y = datos2[0:,0], datos2[0:,1]
plt.scatter(x,y)
```

```python
def plfunc(x,a,b):
    return a*x**b
```

```python
x, y = datos2[0:,0], datos2[0:,1]
plt.scatter(x,y)
fitpars, covmat = curve_fit(plfunc,x,y)
plt.plot(x, plfunc(x, fitpars[0], fitpars[1]), 'r-')
```

```python
covmat
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
xteo = fitpars[0]*x**fitpars[1]
```

```python
R2(y,xteo)
```

```python
x, y = datos1[0:,0], datos1[0:,1]
plt.scatter(x,y)
fitpars, covmat = curve_fit(expfunc,x,y)
yteo = fitpars[0]*e**(x-fitpars[1])
plt.plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-')
plt.xlabel("$X$", fontsize=30, color='g')
plt.ylabel("$Y$", fontsize=30, color='g')
plt.title("Grafica datos 1", fontsize=30, color ='b')
print "La ecuacion es " + str(fitpars[0]) + "*e**(x-" + str(fitpars[1]) + ")"
print "El coeficiente de correlacion es " + str(R2(y, yteo))
```

```python
x, y = datos2[0:,0], datos2[0:,1]
plt.scatter(x,y)
fitpars, covmat = curve_fit(plfunc,x,y)
yteo = fitpars[0]*x**(fitpars[1])
plt.plot(x, plfunc(x, fitpars[0], fitpars[1]), 'r-')
plt.xlabel("$X$", fontsize=30, color='g')
plt.ylabel("$Y$", fontsize=30, color='g')
plt.title("Grafica datos 2", fontsize=30, color ='b')
print "La ecuacion es " + str(fitpars[0]) + "*x**" + str(fitpars[1])
print "El coeficiente de correlacion es " + str(R2(y, yteo))
```

```python
def abssinfunc(datos,a,b,c):
    return a*abs(np.sin(b*(datos - c)))
```

```python
def sinfunc(datos,a,b,c,d):
    return a*np.sin((4*b)*(datos - c))+d

x,y = datos4[0:,0], datos4[0:,1]
fitpars, covmat = curve_fit(sinfunc,x,y)
plt.scatter(x,y)
yteo = sinfunc(x,fitpars[0],fitpars[1], fitpars[2], fitpars[3])
plt.plot(x, yteo, 'r-')
plt.xlabel("$X$", fontsize=30, color='g')
plt.ylabel("$Y$", fontsize=30, color='g')
plt.title("Grafica datos 4", fontsize=30, color ='b')
print "La ecuacion es " + str(fitpars[0]) + "*sin(" + str(4*fitpars[1]) + "*x -" + str(fitpars[2]) + ") + " + str(fitpars[3])
print "El coeficiente de correlacion es " + str(R2(y, yteo))
```
