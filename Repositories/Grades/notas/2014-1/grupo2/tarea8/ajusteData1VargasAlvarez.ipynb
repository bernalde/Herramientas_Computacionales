```python
%pylab inline
from scipy.optimize import curve_fit as cfit
```

```python
a = loadtxt("data1.dat.1")
```

```python
k = a[:,0]
j = a[:,1]
scatter(k, j)
```

```python
figure(figsize=(10, 10))
x = linspace(0, 10, 100)
def poli(x,a,b, c):
    return a*x**b + c
fit, covmat = cfit(poli,x,j)

plot(x, poli(x, fit[0], fit[1], fit[2]), 'r-')

scatter(k, j)
fit[0], fit[1], fit[2]
```

$y=(1.13*10^-⁷)x⁽¹¹^.⁰³⁾+14.18$

```python
def R2(exp, teo):
    err = sum((teo-exp)**2)
    mean = sum(exp)/len(exp)
    var = sum((mean-exp)**2)
    r2 = 1 - err/var
    return r2
```

```python
yteo = poli(x, fit[0], fit[1], fit[2])
yexp = j
r2data1 = R2(yexp, yteo)
r2data1
```
