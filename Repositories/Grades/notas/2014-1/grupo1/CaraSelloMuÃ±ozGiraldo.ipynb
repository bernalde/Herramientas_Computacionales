```python
def moneda1(N_lanzamientos, M_mediciones):
    listaCara=[]
    listaSello=[]
    listaTotal=[]
    f=[]
    k=0.0
    while k<(M_mediciones*1000):
        k+=1000
        j=0.0
        f.append(k)
    while j<M_mediciones:
        sumaCara=0.0
        sumaSello=0.0
        j+=1
        i=0.0
        while i<N_lanzamientos:
            i+=1
            x = np.random.randint(0, 2)
            if x==0:
                sumaCara+=1 
            if x==1:
                sumaSello+=1 
        listaCara.append(sumaCara)
        listaSello.append(sumaSello)
        promC=np.mean(listaCara)
        promS=np.mean(listaSello)
        y=abs(promC-promS)
        listaTotal.append(y)
            
    
    return listaTotal, f   
```

```python
a,b=moneda1(10000, 40)
```

```python
def moneda2(N_lanzamientos, M_mediciones):
    k=0.0
    listaCara=[]
    listaSello=[]
    listaTotal=[]
    f=[]
    while k<(M_mediciones*1000):
        k+=1000
        j=0.0
        f.append(k)
    while j<M_mediciones:
        sumaCara=0.0
        sumaSello=0.0
        j+=1
        i=0.0
        while i<N_lanzamientos:
            i+=1
            x = np.random.randint(0, 2)
            if x==0:
                sumaCara+=1 
            if x==1:
                sumaSello+=1 
        listaCara.append(sumaCara)
        listaSello.append(sumaSello)
        promC=np.mean(listaCara)
        promS=np.mean(listaSello)
        y=abs((float(promC))/(float (promS)))
        listaTotal.append(y)
    
    return listaTotal, f
```

```python
lista2, r=moneda2(5000, 20)
```
