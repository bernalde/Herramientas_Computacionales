```python
x = np.random.rand(100000)*2 - 1
y = np.random.rand(100000)*2 - 1
```

```python
figure(figsize=(10, 10))
scatter(x, y, alpha=0.2, s=10, c='g')
```

```python
ecuacion = np.where(x**2+y**2<1)
```

```python
numX = x[ecuacion]
numY = y[ecuacion]
```

```python
figure(figsize=(10,10))
scatter(x,y, c='b')
scatter(numX,numY, c='g')
```

```python
X = len(numX)  # Numero de puntos dentro del circulo
Y = len(x)   # Numero de puntos totales. 
```

```python
pi = 4* float(X)/float(Y)
print "pi = ", pi
```

```python
x1 = np.random.rand(100000)*6 -3
y1 = np.random.rand(100000)
figure(figsize=(10, 10))
scatter(x1, y1, alpha=0.2, s=10, c='r')

```

```python
def expfunc(x1):
    return (e**-x1**2)
graf = linspace(-3, 3,100000)
figure(figsize=(10, 10))
plot(graf, expfunc(graf), 'r')
```
