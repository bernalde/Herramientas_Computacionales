```python
x = np.random.rand(100000)*6 - 3
y = np.random.rand(100000)
```

```python
figure(figsize=(9, 9))
scatter(x, y, s=10, color='g')
```

```python
index = np.where(y < e**(-x**2))
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(9,9))
scatter(x,y, c='gb')
scatter(xc,yc, c='rmybc')
```

```python
X = len(x)
Y = len(yc)  
```

```python
area = (float(Y)/float(X))*6
print area
```
