```python
def funcion (x):
    y=e**(-x**2)
    return y
```

```python

figure(figsize=(9,9))
x1 = np.random.rand(100000)*8 -4
y1 = np.random.rand(100000)


x = linspace(-4, 4, 50)
y=funcion(x)
plot(x,y,c="r")
scatter(x1,y1,alpha=0.2,s=10)

```

```python
index = np.where(y1<funcion(x1))

```

```python
xc = x1[index]
yc = y1[index]
```

```python
figure(figsize=(9,9))
scatter(x1,y1, c='g')
scatter(xc,yc, c='r')
```

```python
xmax=max(xc)
xmin=min(xc)
```

```python
print xmax
print xmin
```

```python
figure(figsize=(9,9))
scatter(x1,y1, c='g')
scatter(xc,yc, c='r')
axvline(xmax)
axvline(xmin)
```
