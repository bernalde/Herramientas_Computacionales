```python
x = np.random.rand(10000)*6 - 3
y = np.random.rand(10000) 
figure(figsize=(9, 9))
scatter(x, y, alpha=0.2, s=10)
```

```python
def expfunc(x):
    return e**(-x**2)
xa = linspace(-3, 3, 1000)
figure(figsize=(15, 15))
plot(xa, expfunc(xa), 'r-')
```

```python
index = np.where(y<e**(-x**2))
xc = x[index]
yc = y[index]
figure(figsize=(9,9))
scatter(x,y, c='g')
scatter(xc,yc, c='r')
```

```python
X = len(xc)  
Y = len(x) 
```

```python
Area=(float(X)/float(Y))*6
print Area
```
