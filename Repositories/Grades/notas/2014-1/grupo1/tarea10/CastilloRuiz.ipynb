```python
%pylab inline
import numpy as np
```

```python
#puntos  aleatorios 
x = np.random.rand(10000)*6- 3
y = np.random.rand(10000) 
figure(figsize=(10,10))
scatter(x,y, color='c', s=7)
```

```python
xc= np.linspace(-3.5,3.5,10000)
yc= exp(-xc**2)

figure(figsize=(9,9))
plot(xc , yc, label="y=e^(-x^2)")
legend(loc=2)
```

```python
puntosCurva = np.where(y<exp(-x**2))
xg = x[puntosCurva]
yg = y[puntosCurva]
```

```python
figure(figsize=(9,9))
scatter(x, y, c='c', alpha=0.2)
scatter(xg, yg, c='y')
title("$Area\ a\ definir\ por\ metodo\ Monte\ Carlo$", fontsize=20)
```

```python
X= len(xg)#  numero de puntos en el  area de la curva
Y=len(x) # numero de puntos totales 
relac= float(X)/float(Y)

print X, Y, relac
```

```python
Area=relac*6

print Area
```
