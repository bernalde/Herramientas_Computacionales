```python
x1 = np.random.rand(10000)*8 - 4
y1 = np.random.rand(10000)
```

```python
def function(x):
    y=e**(-x**2)
    return y
```

```python

x=linspace(-4, 4, 50)
y=function(x)

scatter(x1, y1, alpha=0.2, c='g', s=10)
plot(x, y, c='r')
```

```python
index = np.where(y1<function(x1))
```

```python
xc = x1[index]
yc = y1[index]
```

```python
figure(figsize=(9,9))
scatter(x1,y1, c='g')
scatter(xc,yc, c='r')
```

```python
min=amin(xc)
max=amax(xc)
print max
print min
```

```python

def No():
    No=0
    i=0
    while (i<(len(x1))):
        if((x1[i]>min) & (x1[i]<max)):
            No+=1
        i+=1
    return No
print No()
Nc=(len(xc))
print Nc
```

```python
def Area():
    A=(float(Nc)/float(No()))*(max-min)
    return A
```
