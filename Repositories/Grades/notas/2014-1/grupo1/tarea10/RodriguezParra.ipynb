```python
xx = np.random.rand(10000)*6 - 3
yy = np.random.rand(10000)
```

```python
figure(figsize=(9, 9))
scatter(xx, yy, alpha=0.5, s=10, c='B')
```

```python
index = np.where(yy < (e**(-xx**2)))
```

```python
xc = xx[index]
yc = yy[index]
```

```python
figure(figsize=(9,9))
scatter(xx,yy, c='g')
scatter(xc,yc, c='r')
```

```python
Q = len(yy)
W = len(yc)
```
