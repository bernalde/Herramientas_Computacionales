```python
y = np.random.rand(10000)
x = np.random.rand(10000)*6 - 3
figure(figsize(9,9))
scatter(x,y)
```

```python
gauss = np.where(y<np.exp(-x**2))
```

```python
xg = x[gauss]
yg = y[gauss]
```

```python
figure(figsize(9,9))
scatter(x, y, c='g')
scatter(xg, yg, c='r')
```

```python
xmax = max(xg)
xmin = min(xg)
xc = []
for i in range (0,len(x)):
    if(x[i]<=xmax and x[i]>=xmin):
        xc.append(x)

X=len(xg)
Y=len(xc)
ratio = float(X)/(Y)
area = ratio*(xmax-xmin)
print area
pi = area**2
print pi
```
