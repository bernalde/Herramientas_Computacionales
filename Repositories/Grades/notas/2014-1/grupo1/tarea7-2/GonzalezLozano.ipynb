```python
%pylab inline
def moneda(N,M):
    listaCara= []
    listaSello= []
    j=0
    while j < M:
         j+=1
         i=0
         sumaCara= 0
         sumaSello= 0
         while i < N:
               i+=1
               a = np.random.randint(0, 2)
               if a== 0:
                  sumaCara+=1
               elif a==1:
                  sumaSello+=1
         fraccionCara= sumaCara/N
         fraccionSello= sumaSello/N
         listaCara.append(fraccionCara)
         listaSello.append(fraccionSello)
    return listaCara, listaSello
          
                
```

```python
fraccionCara, fraccionSello =moneda(1000.0, 1000.0)
fracciones=fraccionCara
print fracciones
```

```python
def media(lista):
    sum=0.0
    for i in range(0,len(lista)):
        sum=sum+lista[i]
 
    return sum/len(lista)
```

```python
def confianza(p):
    listaCara, listaSello= moneda(100.0, 100.0)
    mu= media(listaCara)
    sigma= np.std(listaCara)
    minimo= mu-sigma
    maximo= mu+sigma
    numero=1
    while p< minimo or p>maximo: 
          numero+=1
          minimo= minimo-sigma
          maximo= maximo+sigma
    return numero
```

```python
print confianza(0.45)
print confianza (0.51)
print confianza(0.53)
print confianza (1.7)
```
