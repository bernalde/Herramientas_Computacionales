```python
def lanzamiento(n,m):
    a=0
    b=0
    r=0.0
    p = []
    c=0.0
    s=0.0
    while b<m:
        a=0.0
        while a<n:
            l =numpy.random.randint(1,3)
            if l==1:
                c+=1
            else:
                s+=1
            a+=1
        r=c/n
        p.append(r)
        c=0.0
        s=0.0
        b+=1
    return p
```

```python
tarea = lanzamiento(1000,50000)
```

```python
hist(tarea, bins=20)
axvline(np.std(tarea)+np.mean(tarea),c='r',linewidth=2.5)
axvline(-np.std(tarea)+np.mean(tarea),c='r',linewidth=2.5)
axvline(2*np.std(tarea)+np.mean(tarea),c='y',linewidth=2.5)
axvline(-2*np.std(tarea)+np.mean(tarea),c='y',linewidth=2.5)
axvline(3*np.std(tarea)+np.mean(tarea),c='g',linewidth=2.5)
axvline(-3*np.std(tarea)+np.mean(tarea),c='g',linewidth=2.5)
axvline(4*np.std(tarea)+np.mean(tarea),c='b',linewidth=2.5)
axvline(-4*np.std(tarea)+np.mean(tarea),c='b',linewidth=2.5)
```

```python
def intervalo(v):
    e=np.std(tarea)
    f=np.mean(tarea)
    i=1
    r=0
    while r==0:
        s=(e*i)+f
        if  v<=s:
            r=i
        i+=1
    return "intervalo de contianza sigma",r
```

```python
def calculoValorAbsoluto(Lmin,Lmax,M):
    L=Lmin
    ll=[]
    w=[]
    while L<(Lmax+1):
        ll.append(L)
        a=0
        b=0
        mean=0.0
        p = []
        c=0.0
        s=0.0
        while b<M:
            a=0.0
            while a<L:
                l =numpy.random.randint(1,3)
                if l==1:
                    c+=1
                else:
                    s+=1
                a+=1
            d=abs(c-s)
            c=0.0
            s=0.0
            p.append(d)
            d=0.0
            c=0.0
            b+=1
        mean=np.mean(p)
        w.append(mean)
       
        L+=1000
    return ll, w
```

```python
def calculoRazon(Lmin,Lmax,M):
    L=Lmin
    ll=[]
    w=[]
    while L<(Lmax+1):
        ll.append(L)
        a=0
        b=0
        ratio=0.0
        p = []
        c=0.0
        s=0.0
        d=0.0
        while b<M:
            a=0.0
            while a<L:
                l =numpy.random.randint(1,3)
                if l==1:
                    c+=1
                else:
                    s+=1
                a+=1
            d=(c/s)
            c=0.0
            s=0.0
            p.append(d)
            d=0.0
            c=0.0
            b+=1
        ratio=np.mean(p)
        w.append(ratio)
       
        L+=1000
    return ll, w
```

```python
def calculoDesviacionRazon(Lmin,Lmax,M):
    L=Lmin
    ll=[]
    w=[]
    while L<(Lmax+1):
        ll.append(L)
        a=0
        b=0
        deviation=0.0
        p = []
        c=0.0
        s=0.0
        d=0.0
        while b<M:
            a=0.0
            while a<L:
                l =numpy.random.randint(1,3)
                if l==1:
                    c+=1
                else:
                    s+=1
                a+=1
            d=(c/s)
            c=0.0
            s=0.0
            p.append(d)
            d=0.0
            b+=1
        deviation=np.std(p)
        w.append(deviation)
       
        L+=1000
    return ll,w
```

Entre mayor es el numero de muestras, se vuelve menor la desviacion estandar para los datos de la fraccion y mayor la de la diferencia.
Esto demuestra que a mayor es la cantidad de datos hay una tendecia a que la cantidad de caras sobre con respecto a sellos sea parecida, 
aun asi debido a que hay muchos mas datos aunque las fracciones sean parecidas las diferencias son mucho mas grandes pues hay mas datos:
257/420 (0.611) se parece a 247/430 (0.57) mas que 2/4 (0.5) a 1/5 (0.2) pero 163 se parece menos a 183 que 2 a 4.

Para la fraccion en general el numero de muestras no altera mucho la tendencia, mientras que para la diferencia los datos son mucho menos
irregulares cuando hay mas medidas.
