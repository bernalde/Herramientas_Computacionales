```python
def moneda(lanzamientos,repeticiones):
    i=0
    frac=0.0
    lista=[]
    numerommediciones=[]
    while i<repeticiones:
        p=0
        i+=1
        ca=0.0
        se=0.0
        while p<lanzamientos:
            p+=1
            x= np.random.randint(1,3)
            if x==1:
                ca+=1
            else:
                se+=1
        frac=ca/(ca+se)
        lista.append(frac)
    return lista
```

```python
lista=moneda(200.0,10000.0)
```

```python
hist(lista,bins=12)
title("Histograma fraccion lanzamiento cara", fontsize=18)
xlabel (" Resultados lanzamiento cara")
```

```python
listaAbs=[]
numeromediciones=[]
def moneda2(lanzamientos,mediciones):
    i=0
    absoluto=0.0
    lanza=lanzamientos
    listaAbs=[]
    numeromediciones=[]
    while i<mediciones:
        p=0
        i+=1
        ca=0.0
        se=0.0
        while p<lanza:
            p+=1
            x= np.random.randint(1,3)
            if x==1:
                ca+=1
            else:
                se+=1
        absoluto=abs(ca-se)
        listaAbs.append(absoluto)
        numeromediciones.append(p)
        lanza+=1000
    return (listaAbs, numeromediciones)
```

```python
ejex=[]
ejey=[]
for i in range(1):
    x,y = moneda2(1000,20)
    ejex.append(x)
    ejey.append(y)
```

```python
data =scatter(ejey,ejex)
title("Diferencia de Caras, 20 intentos", fontsize=18)
xlabel (" Lanzamientos")
```
