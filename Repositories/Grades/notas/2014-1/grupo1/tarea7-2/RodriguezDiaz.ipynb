```python
def monedas(N, M):
    i =0
    fracciones = []
    while i < M:
        j=0
        i+=1
        cara=0
        while j< N:
            j+=1 
            a= np.random.randint(1, 3)
            if (a ==1):
                cara+=1  
        fr= cara/N
        fracciones.append(fr)
                
    return fracciones
            
```

```python
def diferencia (N,M):
    i=0
    listacaras= []
    listasellos= []
    while i<M:
        j=0
        i+=1
        caras=0
        sellos=0
        while j < N:
            j+=1
            x = np.random.randint(1, 3)
            if x==1:
                caras+=1
            if x==2:
                sellos+=1
        listacaras.append(caras)
        promediocaras =np.mean(listacaras)
        listasellos.append(sellos)
        promediosellos =np.mean(listasellos)
        
        diferencias = abs(promediocaras-promediosellos)
        
    return diferencias
        
            
```

```python
def puntosdiferencia(M):
    i=1000
    puntos =[]
    nlanz=[]
    while i< 20001:
        i+=1000
        diferencias=diferencia(i,M)
        puntos.append(np.mean(diferencias))
        nlanz.append(i)
    return nlanz, puntos
```

```python
def razon(N,M):
    i=0
    listacaras= []
    listasellos= []
    while i<M:
        j=0
        i+=1
        caras=0
        sellos=0
        while j < N:
            j+=1
            x = np.random.randint(1, 3)
            if x==1:
                caras+=1
            if x==2:
                sellos+=1
        listacaras.append(caras)
        promediocaras =np.mean(listacaras)
        listasellos.append(sellos)
        promediosellos =np.mean(listasellos)
      
        razon = promediocaras/promediosellos  
        
    return razon
```

```python
def puntosrazon(M):
    i=1000
    puntos=[]
    lanzamientos =[]
    while i<20001:
        i+=1000
        razones= razon(i, M)
        puntos.append(np.mean(razones))
        lanzamientos.append(i)
    return lanzamientos, puntos
    
            
        
```

```python
x, y = puntosrazon(20)
a, b = puntosrazon(40)
plt.scatter(x,y)
plt.scatter(a, b, c='red')
plt.title('Razones')
plt.xlabel("Numero de lanzamientos")
plt.ylabel("Razones de cara-sellos")

#En terminos generales se observa que la razon entre cara y sello es muy cercana, es decir, segun el numero de lanzamientos, los valores para cara y sello obtenidos son casi iguales.# 
#Sin embargo, los puntos muy alejados entre sì, van a significar una desviacion de datos significativa#
```

```python
x, y = puntosrazon(20)
a, b = puntosrazon(40)
plt.scatter(x,y)
plt.scatter(a, b, c='r')
plt.title('Razones')
plt.xlabel("Numero de lanzamientos")
plt.ylabel("Razones de cara-sellos")
plt.xscale('log')
plt.yscale('log')

#Los lanzamientos estan en un intervalo de 10 a la 0, hasta un poco mas de 10 a la 4 ya que el valor maximo de lanzamientos es 2500, mientras que las razones estan en un solo orden de magnitud. Es por esto que la grafica obtenida es una "linea recta" de puntos# 
```

```python
def promediodiferencia (N, M):
    i=0
    dife = []
    while i<M:
        j=0
        i+=1
        caras=0
        sellos=0
        while j < N:
            j+=1
            x = np.random.randint(1, 3)
            if x==1:
                caras+=1
            if x==2:
                sellos+=1
        dif = abs(caras-sellos)
        dife.append(dif)
        promdif = mean(dife)
    return promdif
```

```python
def promediodiferencias(M):
    i=1000
    puntos =[]
    nlanz=[]
    while i< 20001:
        i+=1000
        diferencias=promediodiferencia(i,M)
        puntos.append(diferencias)
        nlanz.append(i)
    return nlanz, puntos
```

```python
x, y = promediodiferencias(20)
a, b = promediodiferencias(40)
plt.scatter(x, y)
plt.scatter(a, b, c='red')
plt.title('Promedio diferencia')
plt.xlabel("Numero de lanzamientos")
plt.ylabel("Diferencias de cara-sellos")
plt.xscale('log')
plt.yscale('log')

#Para esta grafica debido a que el numero de lanzamientos aumenta,ello implica que la probabilidad que existe entre la diferencia de cara y sello, aumente. Siendo asi el promedio de las diferencias cada vez mayor#
```

```python
def razones(N,M):
    i=0
    razones =[]
    while i<M:
        j=0
        i+=1
        caras=0
        sellos=0
        while j < N:
            j+=1
            x = np.random.randint(1, 3)
            if x==1:
                caras+=1
            if x==2:
                sellos+=1
        raz = caras/sellos
        razones.append(raz)
        promrazones = mean(razones)
              
    return promrazones
```

```python
def puntosrazones(M):
    i=1000
    puntos=[]
    lanzamientos =[]
    while i<20001:
        i+=1000
        ra= razones(i, M)
        puntos.append(ra)
        lanzamientos.append(i)
    return lanzamientos, puntos
    
```

```python
x, y = puntosrazones(20)
a, b = puntosrazones(40)
plt.scatter(x,y)
plt.scatter(a, b, c='red')
plt.title('Promedio razones')
plt.xlabel("Numero de lanzamientos")
plt.ylabel("Razones de cara-sellos")
plt.xscale('log')
plt.yscale('log')

#La grafica no presenta una forma funcional deducible a partir de la escala logaritmica. Es un patron aleatorio"
```
