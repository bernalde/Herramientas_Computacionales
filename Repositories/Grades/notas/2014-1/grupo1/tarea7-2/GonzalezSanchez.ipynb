```python
def moneda(N_lanzamientos, M_mediciones):
    i=0
    totalCaras = []
    while(i < M_mediciones):
        i+=1       
        caras = 0 
        j=0
        while( j < N_lanzamientos):
            j+=1
            lanzamiento = np.random.randint(0,2)
            if (lanzamiento == 0):
                caras += 1/float(N_lanzamientos)
        totalCaras.append(caras)
    return totalCaras
    
```

```python
prueba = moneda(10000, 10000)
data=hist(prueba, bins = 100)
```

```python
desv = np.std(prueba)
media = np.mean(prueba)
#Se calculan los valores de la media y la desviación estándar
interv1 = (media-0.1)/desv
interv2 = (media-0.35)/desv
interv3 = (media-0.4)/desv
interv4 = (1.7-media)/desv
#Los intervalos de confianza se calculan dividiendo la diferencia entre la media y el valor nominal entre la desviación estándar
print 'Intervalos de confianza:'
print 'Para cada uno de los siguientes valores, se escribe después de cada uno el número de desviaciones estándar a las cuales se encuentra'
print '0.1'
print interv1
print '0.35'
print interv2
print '0.4'
print interv3
print '1.7'
print interv4
```

```python
def comparativa(N_lanzamientos, M_mediciones):
    lanzamientos = []
    CarasSellos = [];
    Caras = []
    Sellos  = []
    i=0
    while (i<M_mediciones):
        i+=1
        caras=0
        sellos=0
        j=0
        while(j<N_lanzamientos):
            j+=1
            lanzamiento = np.random.randint(0,2)
            if(lanzamiento == 0):
                caras += 1
            elif(lanzamiento == 1):
                sellos += 1  
        Caras.append(caras)
        Sellos.append(sellos)
        
        carasp = mean(Caras)
        sellosp = mean(Sellos)
        diferencia = abs(carasp-sellosp)
        CarasSellos.append(diferencia)
        lanzamientos.append((i+1)*1000)
            
    return CarasSellos, lanzamientos
```

```python
y1, x1 = comparativa(1000,20)
y2, x2 = comparativa(1000, 40)
figure(figsize=(9,9))
fig = plt.figure()
fig1 = fig.add_subplot(111)
fig1.scatter(x1,y1, marker='x')
ylabel("Valor absoluto de la diferencia entre caras y sellos")
xlabel("Numero de lanzamientos")
#title("Gráfica 1")
fig1.scatter(x2,y2, marker='^')
```

```python
"""Los resultados reflejan el comportamiento esperado para los lanzamientos sucersivos de una moneda. Entre más +
iteraciones se hagan, más se acercará a cero"""
```

```python
def division(N_lanzamientos, M_mediciones):
    lanzamientos = []
    CarasSellos = [];
    Caras = []
    Sellos  = []
    i=0
    while (i<M_mediciones):
        i+=1
        caras=0
        sellos=0
        j=0
        while(j<N_lanzamientos):
            j+=1
            lanzamiento = np.random.randint(0,2)
            if(lanzamiento == 0):
                caras += 1
            elif(lanzamiento == 1):
                sellos += 1  
        Caras.append(caras)
        Sellos.append(sellos)
        
        carasp = mean(Caras)
        sellosp = mean(Sellos)
        razon = carasp/sellosp
        CarasSellos.append(razon)
        lanzamientos.append((i+1)*1000)
            
    return CarasSellos, lanzamientos
```

```python
y1, x1 = division(1000,20)
y2, x2 = division(1000, 40)
figure(figsize=(9,9))
fig = plt.figure()
fig1 = fig.add_subplot(111)
fig1.scatter(x1,y1, marker='x')
ylabel("Razon entre cara y sello")
xlabel("Nomero de lanzamientos")
#title("Gráfica 1")
fig1.scatter(x2,y2, marker='^')
```

```python
"""Los resultados obtenidos en este caso se comportan como lo esperado para los lanzamientos de una moneda, siendo la
razon entre los promedios de las caras y los sellos convergente a uno"""
```
