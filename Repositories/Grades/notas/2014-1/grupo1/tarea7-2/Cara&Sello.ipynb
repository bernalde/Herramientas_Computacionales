```python
def rango(N_lanzamientos,M_mediciones):
    listaTotal=[]
    listaCara=[]
    listaSello=[]
    final=[]
    k=0.0
    while k<(M_mediciones*1000):
        k+=1000
        m=0.0
        final.append(k)
    while (m < M_mediciones):
        m+=1
        n=0.0
        sumaCara=0
        sumaSello=0
        while (n < N_lanzamientos):
            n+=1
            x = np.random.randint(0, 2)
            if (x==0):
                sumaCara+=1
            if (x==1):
                sumaSello+=1
        listaCara.append(sumaCara)
        listaSello.append(sumaSello)
        pCara= np.mean(listaCara)
        pSello=np.mean(listaSello)
        y=abs(float(pCara)/float(pSello))
        listaTotal.append(y)
        
   
    return listaTotal, final
```

```python
def media(N_lanzamientos,M_mediciones):
    listaTotal=[]
    final=[]
    m=0.0
    while (m < M_mediciones):
        m+=1
        n=0.0
        sumaCara=0
        sumaSello=0
        while (n < N_lanzamientos):
            n+=1
            x = np.random.randint(0, 2)
            if (x==0):
                sumaCara+=1
            if (x==1):
                sumaSello+=1
            listaTotal.append(np.mean(sumaCara-sumaSello))
            final.append(n)
    return listaTotal, final
```

```python
lista3, g = media(200,30)
```

```python
plt.scatter(g, lista3)
xlabel("$Lanzamiento$", fontsize=22)
ylabel("$Media$", fontsize=22)
xscale("log")
yscale("log")
```

```python
def m2(N_lanzamientos,M_mediciones):
    listaTotal=[]
    final=[]
    m=0.0
    k=0.0
    while (m < M_mediciones):
        m+=1
        n=0.0
        sumaCara=0
        sumaSello=0
        while (n < N_lanzamientos):
            n+=1
            x = np.random.randint(0, 2)
            if (x==0):
                sumaCara+=1
            if (x==1):
                sumaSello+=1
            if(sumaSello!=0):
                listaTotal.append(np.mean(float(sumaCara)/float(sumaSello)))
                k+=1
                final.append(k)
    return listaTotal, final
```

```python
def mediast(N_lanzamientos,M_mediciones):
    listaTotal=[]
    final=[]
    m=0.0
    while (m < M_mediciones):
        m+=1
        n=0.0
        sumaCara=0
        sumaSello=0
        while (n < N_lanzamientos):
            n+=1
            x = np.random.randint(0, 2)
            if (x==0):
                sumaCara+=1
            if (x==1):
                sumaSello+=1
            listaTotal.append(np.std(np.mean(sumaCara-sumaSello)))
            final.append(n)
    return listaTotal, final
```
