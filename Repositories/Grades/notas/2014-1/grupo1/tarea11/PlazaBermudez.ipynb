```python
def funcion(x):
    y = (exp(-(x**2)))/(((x-3)**2)+((0.01)**2))
    return y
  
```

```python
figure(figsize=(9,9))
X = linspace(-5, 5,100)
Y = funcion(X)
plot(X,Y)
```

```python
walk = []
x = random.random()*10-5
walk.append(x)
```

```python
for i in range(200000):
    x=random.random()*2-1
    a = funcion(x + walk[-1])/function(walk[-1])
    
    if a>=1.0:
        walk.append(walk[-1]+x)
    else:
        b = random.random()
        if(b<=a):
            walk.append(walk[-1]+x)
        else:
            walk.append(walk[-1])    
```
