```python
x=data4[:,0]
y=data4[:,1]
```

```python
scatter(x,y)
```

```python
def sinfunc(x,l,k,j):
    return np.sin(l*(k**x)+(j*x))**2
fitpars, covmat = curve_fit(sinfunc,x,y)

plot(x, sinfunc(x, fitpars[0], fitpars[1],fitpars[2]), 'r-')
scatter(x, y)
xlabel("$\mathrm{x}$", fontsize=20, color='b')
ylabel("$\mathrm{y}$", fontsize=20, color='b')
title("Regresion", fontsize=30)
```

```python
def R2(exp, teo):
    EE = sum((teo-exp)**2)
    media = sum(exp)/(len(exp))
    vari = sum((media-exp)**2)
    R2 = 1 - EE/vari
    return R2
```

```python
yteo = sinfunc(x, fitpars[0], fitpars[1],fitpars[2])
R2(y, yteo)
```
