```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
data = loadtxt("data3.dat")
```

```python
x = data[:,0]
y = data[:,1]
scatter(x,y, c='g', alpha=0.5, s=20)
xlabel("$\mathrm{x}$", fontsize=20, color='r')
ylabel("$\mathrm{y}$", fontsize=20, color='r')
title("Datos 3", fontsize=30)
```

```python
fit =  polyfit(x, y, 1)

print fit

plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit", color='k')
scatter(x,y, c='g', alpha=0.5, s=20)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25, color='r')
ylabel("$\mathrm{y}$", fontsize=25, color='r')
title("$\mathrm{Regresion\ Lineal}$", fontsize=30)
```

```python
def K(ex, teo):
    E = sum((teo-ex)**2)
    promedio = sum(ex)/(len(ex))
    varianza = sum((promedio-ex)**2)
    K = 1 - E/varianza
    return K
```

```python
yteo  = x*fit[0] + fit[1]
K(y, yteo)
```

```python
data1=loadtxt("data1.dat")
```

```python
q = data1[:,0]
w = data1[:,1]
scatter(q,w, c='g', alpha=0.5, s=20)
xlabel("$\mathrm{x}$", fontsize=20, color='r')
ylabel("$\mathrm{y}$", fontsize=20, color='r')
title("Datos 1", fontsize=30)
```

```python
def expfunc(q,a,b):
    return a*e**(q-b)
fitpars, covmat = curve_fit(expfunc,q,w)

plot(q, expfunc(q, fitpars[0], fitpars[1]), 'r-')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(q, w)
```
