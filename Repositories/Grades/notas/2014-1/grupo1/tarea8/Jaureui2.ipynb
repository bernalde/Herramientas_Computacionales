```python
data3 = loadtxt("DatosTareaReg/data3.dat")
```

```python
x3 = data3[:,0]
y3 = data3[:,1]
```

```python
fit3 =  polyfit(x3, y3, 1)
print fit3
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
y3teo  = x3*fit3[0] + fit3[1]
r2_3 = R2(y3, y3teo)
figure(figsize(9.5,9))
plot(x3, y3teo, linewidth='2.0')
scatter(x3, y3, c='k')
text(7, 1, "$R^2 = $" + str(r2_3), fontsize = 15)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ 3}$", fontsize=25)
```

```python
data1 = loadtxt("DatosTareaReg/data1.dat")
```

```python
x1 = data1[:,0]
y1 = data1[:,1]
```

```python
from scipy.optimize import curve_fit
```

```python
def expfunc(x,a,b):
    return a*e**(x-b)
```

```python
fitpars1, covmat1 = curve_fit(expfunc, x1, y1)
y1teo  = fitpars1[0]*e**(x1 - fitpars1[1])
r2_1 = R2(y1, y1teo)
```

```python
figure(figsize(9.5,9))
plot(x1, y1teo, linewidth='2.0')
scatter(x1, y1, c='k')
text(0, 12000, "$R^2 = $" + str(r2_1), fontsize = 15)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ 1}$", fontsize=25)
```
