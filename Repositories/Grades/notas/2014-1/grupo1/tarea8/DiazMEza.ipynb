```python
data=loadtxt("data3.dat.1")
```

```python
x= data[:,0]
y= data[:,1]
fit=polyfit(x, y, 1)
print fit
plot(x, x*fit[0] + fit[1], label="Fit")
scatter(x, y)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
```

```python
def Rcuadrado(exp, teo):
    EE=sum((teo-exp)**2)
    Prom=(sum(exp)/len(exp))
    Var=sum((Prom-exp)**2)
    R2=1-(EE/Var)
    return R2
    
```

```python
teorico=x*fit[0]+fit[1]
Rcuadrado(y, teorico)
```
