```python
data = loadtxt("data3.dat")
x = data[:,0]
y = data[:,1]
```

```python
scatter(x,y, c='k', s=18)
title("Data3")
xlabel("X", fontsize=18, color='g')
ylabel("Y", fontsize=18, color='g')
```

```python
fit =  polyfit(x, y, 1)

print fit

plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ lineal\ del\ data3}$", fontsize=25)

```

```python
def R(experimental, teorico):
    LS = sum((teorico-experimental)**2)
    promedio = sum(experimental)/(len(experimental))
    DS = sum((promedio-experimental)**2)
    R = 1 - LS/DS
    return R
```

```python
yteorico  = x*fit[0] + fit[1]
R(y, yteorico)
```
