```python
from scipy.optimize import curve_fit
%pylab inline
```

```python
data = loadtxt("data3.dat")
```

```python
x=data[:,0]
y=data[:,1]
```

```python
scatter(x, y)
```

```python
fit =  polyfit(x, y, 1)

print fit

plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$x$", fontsize=20)
ylabel("$y$", fontsize=20)
title("Regresion", fontsize=25)
```

```python
def R2(exp, teo):
    EE = sum((teo-exp)**2)
    media = sum(exp)/(len(exp))
    vari = sum((media-exp)**2)
    R2 = 1 - EE/vari
    return R2
```

```python
yteo  = x*fit[0] + fit[1]
R2(y, yteo)
```
