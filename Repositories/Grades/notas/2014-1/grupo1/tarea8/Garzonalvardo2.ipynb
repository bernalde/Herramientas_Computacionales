```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
data1= loadtxt("data1.dat")
data2= loadtxt("data2.dat")
data3= loadtxt("data3.dat")
data4= loadtxt ("data4.dat")
def Rcuadrado(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
x= data3[:,0]
y= data3[:,1]
scatter(x,y)
fit =  polyfit(x, y, 1)
plot(x, x*fit[0] + fit[1], linewidth='1.5', label="Fit", c="r")
legend(loc='best')
yteo = x*fit[0] + fit[1]
Rcuadrado(y,yteo)
```

```python
x= data1[:,0]
y= data1[:,1]
scatter(x,y)
def expfunc(x,a,b):
    return a*e**(x+2)
fitpars, covmat = curve_fit(expfunc,x,y)
plot(x, expfunc(x, fitpars[0], fitpars[1]), linewidth="1.5",label="fit", c="y")
legend(loc='best')
yteo = expfunc(x, fitpars[0], fitpars[1])
Rcuadrado(y,yteo)
```

```python
x= data2[:,0]
y= data2[:,1]
scatter(x,y)
fit = polyfit(x, y, 2)
plot(x, fit[0]*x**2 + fit[1]*x + fit[2], linewidth=2.5,c="r",label="fit")
yteo = fit[0]*x**2 + fit[1]*x + fit[2]
R2(y, yteo)
```
