```python
%pylab inline
from scipy.optimize import curve_fit
data3= loadtxt("abril/data3.dat")
```

```python
x= data3[:,0]
y= data3[:,1]
scatter(x,y)
```

```python
fit =  polyfit(x, y, 1)
print fit
```

```python
plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Linear\ Regresion}$", size=25) 
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
    
```

```python
yteo  = x*fit[0] + fit[1]
R=R2(y, yteo)
print R
```

```python
figure(figsize(10,10))
plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Linear\ Regresion\ Data3}$", fontsize=25)
text(6,1 ,"$R^2= $"+ str(R), fontsize= 18)
```

```python
#DATA 1
data1= loadtxt("abril/data1.dat")
x1= data1[:,0]
y1= data1[:,1]
scatter(x1,y1)
```

```python
x1= data1[:,0]
y1= data1[:,1]
def expfunc(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(expfunc,x1,y1)

plot(x1, expfunc(x1, fitpars[0], fitpars[1]), linewidth='2.0', label="Fit",c="r")
legend(loc='best')
xlabel("$\mathrm{x}$", size=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(x1, y1, alpha=0.7)
title("$\mathrm{Exponential\ Regresion}$", fontsize=25)
```
