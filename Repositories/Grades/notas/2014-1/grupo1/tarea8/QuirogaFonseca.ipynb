```python
data = loadtxt("data1.dat")
```

```python
x = data[:,0]
y = data[:,1]
```

```python
scatter(x,y)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
```

```python
from scipy.optimize import curve_fit

def expfunc(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(expfunc,x,y)

plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-')
title("Data1")
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(x, y)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
yteo = fitpars[0]*e**(x-fitpars[1])
R2(y, yteo) #Coeficiente de correlación. ¡Funciona!
```
