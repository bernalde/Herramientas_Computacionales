```python
from scipy.optimize import curve_fit
```

```python
data=loadtxt("data3.dat")
```

```python
x=data[:,0]
y=data[:,1]
       
```

```python
scatter(x,y)
```

```python
fit =  polyfit(x, y, 1)

plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Linear\ Regresion}$", fontsize=25)

```

```python
def R2(expe, teor):
    a1=sum((teor-expe)**2)
    promedio=sum(expe)/(len(expe))
    variacion=sum((promedio-expe)**2)
    R2=1-a1/variacion
    return R2
```

```python
yteor= x*fit[0]+fit[1]
R2(y, yteor)

print "La ecuacion de ajuste es:"+ str(fit[0]) +"x"+ str(fit[1])
print "El coeficiente de correlacion es:"+ str(R2(y, yteor))

```

```python
data=loadtxt("data1.dat")
```

```python
x=data[:,0]
y=data[:,1]
```

```python
scatter(x,y)
```

```python
def funcion(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(funcion,x,y)

plot(x, funcion(x, fitpars[0], fitpars[1]))
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(x, y)
```

```python
def R2(expe, teor):
    a1=sum((teor-expe)**2)
    promedio=sum(expe)/(len(expe))
    variacion=sum((promedio-expe)**2)
    R2=1-a1/variacion
    return R2
```

```python
yteor=(fitpars[0]*e**(x-fitpars[1]))
R2(y, yteor)

print "La ecuacion de ajuste es:" + str(fitpars[0]) + "*e**" + ("x" + str(fitpars[1]))
print "El coeficiente de correlacion es:"+ str(R2(y, yteor))
```
