```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
data = loadtxt("data1.dat")
```

```python
x = data[:,0]
y = data[:,1]
scatter(x,y)
```

```python
def expfunc(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(expfunc,x,y)

plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(x, y)
```

```python
figure(figsize(12,12))
plot(x, expfunc(x, fitpars[0], fitpars[1]), linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ Lineal}$", fontsize=25)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
yteo  = expfunc(x, fitpars[0], fitpars[1])
r = R2(y, yteo)
print r
```

```python
figure(figsize(10,10))
plot(x, expfunc(x, fitpars[0], fitpars[1]), linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ Lineal - 1 }$", fontsize=25)
text (0,12000, "$R^2 =$"+ str(r), fontsize =12)
```
