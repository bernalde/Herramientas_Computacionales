```python
data = loadtxt("data3.dat")
```

```python
x = data[:,0]
y = data[:,1]
scatter(x,y)
```

```python
fit  = polyfit(x, y, 1)
print fit
```

```python
plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
plot(x, y, linewidth=2.0, c='y', label="Real")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ Lineal}$", fontsize=25)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
yteo  = x*fit[0] + fit[1]
r = R2(y, yteo)

```

```python
figure(figsize(12,12))
plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ Lineal - 2 }$", fontsize=25)
text (6,1, "$R^2 =$"+ str(r), fontsize =12)
```
