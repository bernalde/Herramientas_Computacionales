```python
scatter(x,y, c='k', s=18)
title("Data3")
xlabel("X", fontsize=18, color='g')
ylabel("Y", fontsize=18, color='g')
```

```python
fit =  polyfit(x, y, 1)

print fit

plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.7)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ lineal\ del\ data3}$", fontsize=25)

```

```python
def R(experimental, teorico):
    LS = sum((teorico-experimental)**2)
    promedio = sum(experimental)/(len(experimental))
    DS = sum((promedio-experimental)**2)
    R = 1 - LS/DS
    return R
```

```python
yteorico  = x*fit[0] + fit[1]
R(y, yteorico)
```

```python
data1 = loadtxt("data4.dat")
m = data1[:,0]
w = data1[:,1]
```

```python
scatter(m,w, c='k', s=18)
title("Data4")
xlabel("X", fontsize=18, color='g')
ylabel("Y", fontsize=18, color='g')
```

```python
from scipy.optimize import curve_fit


```

```python
def senfuc(m,a,b,c):
    return np.sin(a*(b**x)+(c*x))**2
fitpars, covmat = curve_fit(senfuc,m,w)

plot(m, senfuc(m, fitpars[0], fitpars[1], fitpars[2]), 'g-')
scatter(m,w)
xlabel("$\mathrm{x}$", fontsize=15, color='g')
ylabel("$\mathrm{y}$", fontsize=15, color='g')
title("Regresion lineal del Data4", fontsize=20)
```

```python
wteorico  = senfuc(m, fitpars[0], fitpars[1], fitpars[2])
R(w, wteorico)
```

```python
data2 = loadtxt("data2.dat")
a = data2[:,0]
b = data2[:,1]
```

```python
scatter(a,b, c='k', s=18)
title("Data2")
xlabel("X", fontsize=18, color='g')
ylabel("Y", fontsize=18, color='g')
```

```python
def plfunc(a,c,d):
    return c*a**d
fitpars, covmat = curve_fit(plfunc,a,b)

plot(a, plfunc(a, fitpars[0], fitpars[1]), 'g-')
scatter(a, b)
xlabel("$\mathrm{x}$", fontsize=15, color='g')
ylabel("$\mathrm{y}$", fontsize=15, color='g')
title("Regresion lineal del Data2", fontsize=20)
```

```python
bteorico  = plfunc(a, fitpars[0], fitpars[1])
R(b, bteorico)
```

```python
data3 = loadtxt("data1.dat")
f = data3[:,0]
d = data3[:,1]
```

```python
scatter(f,d, c='k', s=18)
title("Data1")
xlabel("X", fontsize=18, color='g')
ylabel("Y", fontsize=18, color='g')
```

```python
def expfunc(f,a,b):
    return a*e**(f-b)
fitpars, covmat = curve_fit(expfunc,f,d)

plot(f, expfunc(f, fitpars[0], fitpars[1]), 'r-')
scatter(f, d)
xlabel("$\mathrm{x}$", fontsize=15, color='g')
ylabel("$\mathrm{y}$", fontsize=15, color='g')
title("Regresion lineal del Data1", fontsize=20)
```

```python
dteorico  = expfunc(f, fitpars[0], fitpars[1])
R(d, dteorico)
```
