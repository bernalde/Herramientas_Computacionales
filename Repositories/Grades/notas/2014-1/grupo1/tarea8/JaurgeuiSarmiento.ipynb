```python
data3 = loadtxt("DatosTareaReg/data3.dat")
```

```python
x = data3[:,0]
y = data3[:,1]
```

```python
fit =  polyfit(x, y, 1)
print fit
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
yteo  = x*fit[0] + fit[1]
r2 = R2(y, yteo)
plot(x, x*fit[0] + fit[1], linewidth='2.0')
scatter(x, y, c='k', alpha=0.7)
text(7, 1, "$R^2 = $" + str(r2), fontsize = 15)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Linear\ Regresion \ 3}$", fontsize=25)
```
