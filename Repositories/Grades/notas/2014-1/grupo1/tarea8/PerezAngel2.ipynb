```python
data = loadtxt("data3.dat")
```

```python
x=data[:,0]
y=data[:,1]
```

```python
scatter(x, y)

```

```python
fit =  polyfit(x, y, 1)

print fit

plot(x, x*fit[0] + fit[1], linewidth='1.0', label="Fit")

scatter(x, y, c='r', alpha=0.7)
legend(loc='best')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ Lineal}$", fontsize=25)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
yteo  = x*fit[0] + fit[1]
R2(y, yteo)
```

```python
data = loadtxt("data1.dat")
```

```python
x=data[:,0]
y=data[:,1]
```

```python
scatter(x, y)
```

```python

from scipy.optimize import curve_fit


```

```python
def expfunc(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(expfunc,x,y)

plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-')
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
scatter(x, y, c='r', alpha=0.7)

xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Regresion\ Exponencial}$", fontsize=25)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
yteo  = expfunc(x, fitpars[0], fitpars[1])
R2(y, yteo)
```
