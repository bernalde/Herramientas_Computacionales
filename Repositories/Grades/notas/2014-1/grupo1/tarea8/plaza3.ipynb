```python
%pylab inline
data = loadtxt("data3.dat")
```

```python
figure(figsize=(9, 6))
x = data[:,0]
y = data[:,1]
scatter(x, y, c='k', alpha=0.7, s=50)
xlabel("x", fontsize=25)
ylabel("y", fontsize=25)
title("Grafica", fontsize=25)
```

```python
figure(figsize=(9, 6))
fit =  polyfit(x, y, 1)
print fit

plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
scatter(x, y, c='k', alpha=0.9)
legend(loc='best')
xlabel("x", fontsize=25)
ylabel("y", fontsize=25)
title("Regresion lineal", fontsize=25)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
yteo  = x*fit[0] + fit[1]
r2 = R2(y, yteo)
```

```python
y1 = y-r2
figure(figsize=(10, 9))
plot(x, x*fit[0] + fit[1], linewidth='2.0', label="Fit")
plot(x, y1, linewidth='2.0', label="real")
scatter(x, y, c='k', alpha=0.9)
legend(loc='best')
text(4, 1, "R2 =" + str(r2),fontsize=20)
xlabel("x", fontsize=25)
ylabel("y", fontsize=25)
title("Regresion lineal(2)", fontsize=25)
```
