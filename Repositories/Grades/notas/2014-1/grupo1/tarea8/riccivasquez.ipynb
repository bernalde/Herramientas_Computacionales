```python
fit =  polyfit(x, y, 1)

m=fit[0]
b=fit[1]
x1 = linspace(0, 10, 100)
y1=m*x1+b
plot(x1,y1)
scatter(x,y)
```

```python
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```
