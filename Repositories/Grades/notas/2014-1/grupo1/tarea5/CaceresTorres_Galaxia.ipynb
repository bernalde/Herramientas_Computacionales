```python
data=loadtxt("data.txt")
figure(figsize(9.5,9))
x=data[:,2]
y=data[:,3]
scatter(x,y)
ylim([30,37])
xlim([15,25])
title("$\mathrm{Grafica\ Galaxia}$",fontsize=35)
scatter(x,y,s=0.5,alpha=0.5,c='k')
tick_params(labelsize=20,colors='g')
xlabel("$x$", fontsize=30)
ylabel("$y$", fontsize=30)
grid(color='k', linestyle='-',linewidth=0.5)





```
