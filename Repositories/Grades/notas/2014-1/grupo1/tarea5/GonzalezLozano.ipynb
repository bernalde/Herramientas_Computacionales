```python
import matplotlib.pyplot as plt
%pylab inline
```

```python
data= loadtxt("data.txt")
```

```python
x = data[:,3]
y = data[:,4]
```

```python
figure(figsize=(15, 15))
scatter(x, y, alpha=0.6)
xlabel("$x$", fontsize=30)
ylabel("$y$", fontsize=30)
plt.xlim(20, 40)
plt.ylim(20, 40)
tick_params(axis="both", labelsize=20, colors="r")
title("La distribucion", fontsize=30)
```
