```python
import matplotlib.pyplot as plt


%pylab inline

```

```python
data = loadtxt("data.txt")
```

```python

figure(figsize=(10, 9))
plt.scatter(x, y, edgecolor='r', facecolor='k', alpha=0.4, s=79)
xlim([19, 23])
ylim([29, 35])
xlabel("$x$", fontsize=35)
ylabel("$y$", fontsize=35)
title("Cluster!", fontsize=30)

```
