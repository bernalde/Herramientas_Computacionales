```python
b=pi*7/16
plt.figure(figsize(10,10))
plt.subplots_adjust(hspace=0.2, wspace =0.2)

matplotlib.rcParams.update({'font.size': 10, 'font.family': 'serif'})

theta1=np.linspace(0,5*2*pi,1000)
r1=theta1
x1=r1*cos(theta1)
y1=r1*sin(theta1)
plt.subplot(223)
plot(x1,y1,'r-.',lw=2.5)
plt.title('Spiral of Archimedes')
plt.ylim([-35,35])
plt.xlim([-30,40])
plt.tick_params(axis='both', labelsize=10, color="k")

theta2=np.linspace(0,5*2*pi,1000)
r2=exp(theta2/tan(b))
x2=r2*cos(theta2)
y2=r2*sin(theta2)
plt.subplot(222)
plot(x2,y2,'b:',lw=2.5)
plt.title('Equiangular Spiral')
plt.ylim([-500,400])
plt.xlim([-300,600])
plt.tick_params(axis='both', labelsize=10, color="k")

theta3=np.linspace(0,15*2*pi,1000)
r3a=sqrt(theta3)
r3b=-sqrt(theta3)
x3a=r3a*cos(theta3)
y3a=r3a*sin(theta3)
x3b=r3b*cos(theta3)
y3b=r3b*sin(theta3)
plt.subplot(221)
plot(x3a,y3a,'b',lw=1)
plot(x3b,y3b,'g',lw=1)
plt.title('Fermat\'s Spiral')
plt.ylim([-10,10])
plt.xlim([-10,10])
plt.tick_params(axis='both', labelsize=10, color="k")

theta4=np.linspace(pi/3,5*2*pi,1000)
r4=1/theta4
x4=r4*cos(theta4)
y4=r4*sin(theta4)
plt.subplot(224)
plot(x4,y4,'^m',ms=3)
plt.title('Hyperbolic Spiral')
plt.ylim([-0.4,1])
plt.xlim([-0.6,0.8])
plt.tick_params(axis='both', labelsize=10, color="k")

plt.savefig("Espirales.pdf", format='pdf', bbox_inches='tight', transparent=True)
```
