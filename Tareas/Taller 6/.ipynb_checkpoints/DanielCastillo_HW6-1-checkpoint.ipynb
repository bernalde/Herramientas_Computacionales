```python
x=np.zeros(0)
y=np.zeros(0)
with open('higgs.csv', 'rb') as csvfile:
     spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
     for row in spamreader:
         x=np.append(x,float(row[0]))
         y=np.append(y,float(row[1]))
```

```python
m=numpy.linspace(105,153,1000)
a0=27890
a1=-526.3
a2=3.410
a3=-0.007502
Ev=a0+a1*m+a2*m*m+a3*m*m*m
c=116
m_star=125
d=4.2
Ev_star=Ev+c*exp(-(m-m_star)*(m-m_star)/d)
```

```python
figure(figsize(8,8))
matplotlib.rcParams.update({'font.size': 15, 'font.family': 'serif'})
plt.errorbar(x,y,yerr=35,c='k',fmt='o',label="Data")
plt.plot(m,Ev,'--r',linewidth=2.5,label="$Ev$")
plt.plot(m,Ev_star,'-b',linewidth=2.5,label="$Ev*$")
p=plt.legend(loc='best')
p.draw_frame(False)

plt.xlabel(r'$m_{\gamma\gamma}$ (GeV)')
plt.ylabel('S/(S+B) Weighted Events / $1.5$ GeV')
plt.title(r'CMS $\sqrt{s}=7$ $\rm{TeV}$, $L=5.1$ $\rm{fb}^{-1}$ $\sqrt{s}=8$ $\rm{TeV}$, $L=5.3$ $\rm{fb}^{-1}$', fontsize=15)
plt.xlim([105,153])
plt.ylim([0,2000])
plt.tick_params(axis='both', labelsize=15, color="k")
plt.minorticks_on()
plt.savefig("CMSHiggs.pdf", format='pdf', bbox_inches='tight', transparent=True)
```
