```python
#b.
def R2(xexp, xteo):
    EE = sum((xteo-xexp)**2)
    mean = sum(xexp)/(len(xexp))
    variance = sum((mean-xexp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
Rsqr=R2(esperanza,esperanzafit)
print 'El coeficiente de correlación (R²) es: ' + str(Rsqr)
```

```python
#David Bernal 200921359 Tarea 10 Punto 2 Herramientas computacionales
import os
import csv
from scipy.optimize import curve_fit
```

```python
#Importar datos
if(os.path.isfile('solemio.csv')):
    os.remove('solemio.csv')
!wget https://raw.githubusercontent.com/jngaravitoc/HerramientasComputacionales/master/Homeworks/HW10/solemio.csv
```

```python
#a.
dia=np.zeros(0)
hora=np.zeros(0)
azimuth=np.zeros(0)
with open('solemio.csv', 'rb') as csvfile:
     spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
     for row in spamreader:
         dia=np.append(dia,float(row[0]))
         hora=np.append(hora,float(row[1]))
         azimuth=np.append(azimuth,float(row[2]))
```

```python
freq=2*pi/365
def myfunky(t,A,B,C):
    return A+B*cos(freq*t)+C*sin(freq*t)
def myfunky2(t,A,B,C,D,E):
    return A+B*cos(freq*t)+C*sin(freq*t)+D*cos(2*freq*t)+E*sin(2*freq*t)
```

```python
fitpars1, covmat = curve_fit(myfunky,dia,azimuth)
plot(dia,myfunky(dia,fitpars1[0],fitpars1[1],fitpars1[2]))
plot(dia,azimuth,'o')
xlabel('Dia')
ylabel('Azimuth')
legend(['Linealizacion','Datos'],0)
grid()
```

```python
#b.
fitpars2, covmat = curve_fit(myfunky2,dia,hora)
plot(dia,myfunky2(dia,fitpars2[0],fitpars2[1],fitpars2[2],fitpars2[3],fitpars2[4]))
plot(dia,hora,'o')
xlabel('Dia del anho')
ylabel('Hora de salida del sol')
legend(['Linealizacion','Datos'],0)
grid()
```

```python
#c.
def myfunkyNR(t):
    A=fitpars2[0]
    B=fitpars2[1]
    C=fitpars2[2]
    D=fitpars2[3]
    E=fitpars2[4]
    return A+B*cos(freq*t)+C*sin(freq*t)+D*cos(2*freq*t)+E*sin(2*freq*t)-6 #menos 6 porque es a las 6 am
    
def myfunkyprime(t):
    A=fitpars2[0]
    B=fitpars2[1]
    C=fitpars2[2]
    D=fitpars2[3]
    E=fitpars2[4]
    return -freq*B*sin(freq*t)+C*freq*cos(freq*t)-2*D*freq*sin(2*freq*t)+2*E*freq*cos(2*freq*t)
```

```python
def NR(d0):
    error=100;
    maxiter=1000;
    iter=0;
    tol=1e-6;
    while (error>tol)&(iter<maxiter):
        t1=d0-myfunkyNR(d0)/myfunkyprime(d0)
        error=abs((t1-d0)/d0)
        iter+=1
        d0=t1
    return d0
```

```python
plot(dia,myfunky2(dia,fitpars2[0],fitpars2[1],fitpars2[2],fitpars2[3],fitpars2[4]))
plot(dia,hora,'o')
xlabel('Dia del anho')
ylabel('Hora de salida del sol')
legend(['Linealizacion','Datos'],0)
grid()
axhline(y=6,c='r')
```

```python
#Nótese que existen 2 días en que esto se cumple y se encuentran gracias a las diferentes inicializaciones
dia1=NR(100)
dia2=NR(350)
print dia1
print dia2
```

Aproximando esos días se obtiene que aquellos días que amanece justo a las 6am son el 26 de marzo (día 85) y el 20 de diciembre (día 384)
