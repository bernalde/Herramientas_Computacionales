```python
figure(figsize=(10, 10))
plt.plot(x,y,c='k')
plt.scatter(x,y,s=10,c='g')
plt.arrow(x[0],y[0],x[-1],y[-1],fc='r',ec='r',head_width=0.5, head_length=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejemplo marcha aleatoria')
plt.grid()
```

```python
def make_walk(N):
    runs=0;
    xi=0.0;
    yi=0.0;
    for i in range(1,N):
        r0=(np.random.random(2)-0.5);
        x0=r0[0];
        y0=r0[1];
        norma=np.sqrt(x0*x0+y0*y0);
        xi=xi+x0/norma;
        yi=yi+y0/norma;
    dist=np.sqrt(xi*xi+yi*yi);
    return dist
```

```python
Nmax=1000;
d=np.zeros(Nmax)
for N in range(1,Nmax):
    D=np.zeros(walks)
    for i in range(walks):
        D[i]=make_walk(N)
    d[N]=np.mean(D)
```

```python
from scipy.optimize import curve_fit
N=np.linspace(0,Nmax,Nmax)
def sqrtfunc(x,a,b):
    return a*np.sqrt(x)+b
fitpars, covmat = curve_fit(sqrtfunc,N,d)
```

```python
figure(figsize=(10, 5))
plt.plot(N, sqrtfunc(N, fitpars[0], fitpars[1]), 'r-')
plt.scatter(N,d,s=0.5)
plt.xlim(0,Nmax)
plt.ylim(0)
plt.ylabel('$<d>$')
plt.xlabel('N')
plt.title('Distancia media contra numero de pasos')
plt.grid()
```
